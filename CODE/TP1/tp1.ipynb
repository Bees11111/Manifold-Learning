### Question 1

```python
# Fonction pour implémenter un générateur congruentiel linéaire
def generateur_congruentiel_lineaire(seed=42, periode=1000, A=5, B=3):
    """Générateur congruentiel linéaire (LCG).

    Arguments :
    - seed : graine initiale
    - periode : nombre de valeurs à générer
    - A : multiplicateur
    - C : incrément

    Retourne :
    - Une liste des nombres pseudo-aléatoires générés
    """

    # Modulo choisi comme une grande valeur (souvent 2^32)
    m = 2**8
    x = seed
    valeurs = []

    for _ in range(periode):
        x = (A * x + B) % m
        valeurs.append(x)

    return valeurs
```

```python
# Exemple d'utilisation
resultat = generateur_congruentiel_lineaire()
print(resultat)
```

### Question 2

```python
# Fonction ajustée pour générer des nombres entre 0 et 1
def generateur_congruentiel_normalise(seed=1, periode=1000, A=43, B=0, m=2048):
    """Génère une suite de valeurs pseudo-aléatoires normalisées entre 0 et 1.
    
    Arguments :
    - seed : graine initiale
    - periode : nombre de valeurs à générer
    - A : multiplicateur
    - C : incrément
    - m : module
    
    Retourne :
    - Une liste de valeurs pseudo-aléatoires normalisées entre 0 et 1.
    """
    x = seed
    valeurs = []
    
    for _ in range(periode):
        x = (A * x + B) % m
        valeurs.append(x / m)  # Normaliser entre 0 et 1
    
    return valeurs
```

```python
# Exemple d'utilisation
resultat_normalises = generateur_congruentiel_normalise()
print(resultat)
```

### Question 3

```python
# Tracer la suite de valeurs normalisées
plt.figure(figsize=(10, 6))
plt.plot(resultat_normalises, marker='o', linestyle='-', markersize=3)
plt.title('Suite de valeurs pseudo-aléatoires normalisées entre 0 et 1')
plt.xlabel('Index')
plt.ylabel('Valeur')
plt.grid(True)
plt.show()
```
