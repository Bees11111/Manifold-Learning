```python
import pandas as pd
from geopy.distance import geodesic
from geopy.geocoders import Nominatim
import numpy as np

# Liste des capitales du monde (10 européennes, 10 africaines, 10 américaines, 10 asiatiques)
cities = [
    # Européennes
    "Paris", "London", "Berlin", "Madrid", "Rome", "Vienna", "Amsterdam", "Brussels", "Copenhagen", "Stockholm",
    # Africaines
    "Cairo", "Nairobi", "Dakar", "Algiers", "Rabat", "Lagos", "Addis Ababa", "Pretoria", "Accra", "Tunis",
    # Américaines (Nord et Sud)
    "Washington D.C.", "Ottawa", "Mexico City", "Buenos Aires", "Brasilia", "Santiago", "Bogota", "Havana", "Lima", "Quito",
    # Asiatiques
    "Tokyo", "Beijing", "New Delhi", "Bangkok", "Seoul", "Jakarta", "Riyadh", "Kuala Lumpur", "Hanoi", "Manila"
]

# Récupération des coordonnées géographiques de chaque ville
geolocator = Nominatim(user_agent="city_distance_calculator")
coordinates = []
for city in cities:
    location = geolocator.geocode(city)
    coordinates.append((location.latitude, location.longitude))

# Calcul des distances en kilomètres entre les villes
distances = np.zeros((len(cities), len(cities)))
for i in range(len(cities)):
    for j in range(i + 1, len(cities)):
        distance = geodesic(coordinates[i], coordinates[j]).km
        distances[i][j] = distance
        distances[j][i] = distance

# Création du DataFrame pandas
distance_df = pd.DataFrame(distances, index=cities, columns=cities)

# Affichage du DataFrame
distance_df

# Impression de la matrice des distances
print(distance_df)
```
